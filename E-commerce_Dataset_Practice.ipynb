```python
df["Discounted_Price"]=(df["Price"]-(df["Price"]*df["Discount_Percent"]/100))
df["Profit_Margin"]=(
    df["Profit"]/df["Revenue"]
)*100
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(8,5))
sns.heatmap(corr,annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.groupby("Category").agg({
    "Revenue":["sum","mean","max"],
    "Profit":["sum","mean"],
    "Rating":["mean"],
    "Delivery_Days":["mean"]
})
```

```python
top_products=df.groupby("Product")["Profit"].sum().sort_values(ascending=False)
print(top_products.head(10))
```

```python
high_profit=df[(df["Profit"]>50000)&
    (df["Rating"]>4.5)]
high_profit.head(10)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Revenue"],bins=30)
plt.title("Revenue_Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df ,x="Revenue", y="Profit", hue="Category")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x="Category", y="Profit")
plt.title("Category with Profit")
plt.show()
```

```python
sns.pairplot(df[["Price","Revenue","Profit","Rating"]])
plt.show()
```

```python
daily_revenue=df.groupby("Delivery_Days")["Revenue"].mean()
daily_revenue.plot(kind="line")
plt.title("Revenue Trend")
plt.show()
```

```python
current_revenue = df["Revenue"].mean()

future_prediction = current_revenue * 1.10

print(future_prediction)
```

```python
df["Moving_Avg"] = (
    df["Revenue"]
    .rolling(window=5)
    .mean()
)

print(df[["Revenue","Moving_Avg"]].head(10))
```

```python
plt.figure(figsize=(12,6))

plt.plot(df["Revenue"].head(100))

plt.plot(df["Moving_Avg"].head(100))

plt.legend(["Revenue","Moving Average"])

plt.show()
```

```python
top_customers = df.groupby(
    "Customer_Name"
)["Revenue"].sum().sort_values(
    ascending=False
)

print(top_customers.head(10))
```
